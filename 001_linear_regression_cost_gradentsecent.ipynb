```python
#building the Model GDA
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import time 
```

```python


def GDA(X,y):
    m_c = b_c =0
    itera =1000
    n = len(X)
    l_r  = 0.001
    for i in range(itera):
        yp = m_c * X + b_c
        cost =(1.0 /n)*sum ([ val ** 2 for val in (y -yp)] )
        md = (-2.0 / n )*sum(X*(y-yp))
        bd = (-2.0/n)*sum(y-yp)
        m_c =m_c - l_r*md
        b_c = b_c - l_r*bd
        m = m_c
        b = b_c
        yp =m*X+b
        plt.scatter(X,y)
        plt.plot(X,yp)
        #time.sleep(0.5)
        #print"m : ",m_c ,"  b_c ",b_c," cost : ",cost
X = np.array([1,3,5,7,9])
y = np.array([1,10,39,52,87])
GDA(X,y)
```
