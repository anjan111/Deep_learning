```python
import numpy as np
X = np.array(([2,9] , [1,5], [3,6],[6,9],[10,8],[10,5],[5,10],[5,13]),dtype = float)
y = np.array(([92],[86],[89],[94],[96],[93],[93],[94]),dtype = float)
print X
print y
# scale the data
X = X/ np.amax(X)
y = y/100
print X
print y
```

```python
class NeuralNetwork():
    def __init__(self):
        # define the parameters
        self.in_size = 2
        self.out_size = 1
        self.hidden_size = 3
        
        self.W1 = np.random.rand(self.in_size,self.hidden_size)
        self.W2 = np.random.rand(self.hidden_size ,self.out_size)
    def sigmoid(self,s,deriv = False):
        if(deriv == True):
            return s*(1-s)
        else:
            return  1/(1+np.exp(-s))
    def feedForward(self,X):
        self.z =np.dot(X,self.W1)
        self.z2 = self.sigmoid(self.z)
        self.z3 = np.dot(self.z2,self.W2)
        output = self.sigmoid(self.z3)
        return output
    def backward(self,X,y,output):
        self.error_out = y - output # error in output
        self.out_delta =self.error_out * self.sigmoid(output ,deriv = True) 
        
        self.z2_error = self.out_delta.dot(self.W2.T)
        self.z2_delta = self.z2_error * self.sigmoid(self.z2, deriv = True)
        
        self.W1  +=X.T.dot(self.z2_delta)
        self.W2  += self.z2.T.dot(self.out_delta)
    def train(self,X,y):
        output= self.feedForward(X)
        self.backward(X,y,output)

NN = NeuralNetwork() 
print str(NN.feedForward(X))
print "***********************"
for i in range(1000):
    if(i % 100 ==0 ):
        print " loss :"+ str(np.mean(np.square(y-NN.feedForward(X))))
    NN.train(X,y)
print " inputs : "+str(X)
print "original  : "+str(y)
print " loss :"+ str(np.mean(np.square(y-NN.feedForward(X)))) 
print
print "final output : "+str(NN.feedForward(X))
```
